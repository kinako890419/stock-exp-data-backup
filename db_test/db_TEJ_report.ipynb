```python
company_ids = pd.read_sql('SELECT id, company_symbol FROM company', con=engine)
# df['company_symbol'] = df['證券代碼'].astype(str)
df['company_symbol'] = df['公司'].astype(str)
company_id_map = dict(zip(company_ids['company_symbol'], company_ids['id']))
df['company_id'] = df['company_symbol'].map(company_id_map)
df
```

```python
indicators = pd.read_sql('SELECT id, tw_indicator FROM singleindicator', engine)
```

```python
indicators['tw_indicator'].values
```

```python
# 財報資料處理

data_to_insert = []
for column in df.columns[:]:
    if column in indicators['tw_indicator'].values:
        indicator_id = indicators[indicators['tw_indicator'] == column].id.values[0]
        for index, row in df.iterrows():
            data_to_insert.append({
                'company_id': row['company_id'],
                'date': row['年/月'],
                # 'date': row['年月日'],
                # 'date': row['date'],
                'period_id': row['period_id'],
                'indicator_id': indicator_id,
                'indicator_value': row[column]
            })
```

```python
data_df = pd.DataFrame(data_to_insert)
data_df
```

```python
# df['年月日'] = pd.to_datetime(df['年月日'])
```

```python
# 將 '年/月' 格式的字符串轉換為 datetime 並設置為當月交易日的最後一天
data_df['date'] = pd.to_datetime(data_df['date'], format='%Y/%m') + pd.offsets.MonthEnd(0)

# 定義一個函數來檢查日期是否為週末，如果是，則移動到最近的工作日
def adjust_to_weekday(date):
    # 如果是星期六，則往前移動一天（星期五）
    if date.weekday() == 5:
        return date - pd.Timedelta(days=1)
    # 如果是星期日，則往前移動兩天（星期五）
    elif date.weekday() == 6:
        return date - pd.Timedelta(days=2)
    # 否則，保持不變
    else:
        return date

# 將調整日期的函數應用到 'date' 欄位
data_df['date'] = data_df['date'].apply(adjust_to_weekday)
```

```python
# 去除空白跟逗號
data_df['indicator_value'] = data_df['indicator_value'].apply(
    lambda x: str(x).strip().replace(',', '') if isinstance(x, str) else x
)

# 轉成浮點樹，N.A. 轉成 NaN
data_df['indicator_value'] = pd.to_numeric(data_df['indicator_value'], errors='coerce')
data_df
```
